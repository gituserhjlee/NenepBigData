```
path = os.listdir("/content/drive/MyDrive/tokenDatas")
def load_data(path):
    movie_review = []
    ma = movie_review.append
    for movie in path:
        data = pd.read_csv("/content/drive/MyDrive/tokenDatas/"+movie, names = ["idx", "review"])
        data = data.iloc[1:,1:]
        review_list = []
        ra = review_list.append
        for i in range(len(data)):
            ra(data.iloc[i,0])
        ma(review_list)
    return movie_review
```

```
token_data=load_data(path)
```

```
model = Word2Vec(token_data[1], #토르 
                sg = 1,
                window = 5,
                min_count = 1
                )
model.init_sims(replace = True)
```
